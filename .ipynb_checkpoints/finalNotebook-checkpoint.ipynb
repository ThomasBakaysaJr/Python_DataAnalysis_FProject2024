```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import scipy as scp

#for headers#
#for both / curated data
ofc = 'Offense Code'
sex = 'Gender'
age = 'Age'
sen = 'Sentence (Years)'
off = 'Offense'
sev = 'Severity' #same as cat, but specifially for the curated data
rac = 'Race'
dat = 'Sentence Date'
fac = 'Current Facility'

#for dates since they put stuff in there
#life
DEATH = 'Death' #Death'
LIFE = 'Life' #Life
CAPLIFE = 'Capital Life' #'Capital Life'
LWOP = 'LWOP' #'LWOP'
#USE by using these as keys for senFilter
#so use like sentence[DEATH]


#lit and sev is for codes
lit = 'Literal'    
cat = 'L/D' #category like 3rd degree felony

#tdo is for data
tdo = 'TDCJ Offense'


#Data I'm not going to use for this analysis
toDrop = ['SID Number', 'Name', 'Maximum Sentence Date', 'Projected Release', 'Parole Eligibility Date', 'Case Number', 'County', 'Last Parole Decision', 'Next Parole Review Date',
       'Parole Review Status', tdo, 'TDCJ Number', 'Offense Date']

def writeOffCode (data, codes):    
    dataOffCode = data[[ofc, tdo]].copy().drop_duplicates(subset=[ofc])
    codeOffCode = codes[[ofc, lit, cat]].copy()
    #combSet = codeOffCode.join(dataOffCode.set_index(oc), on=oc).copy()
    dataOffSer = dataOffCode.drop(tdo, axis='columns')
    combSet = dataOffSer.copy()
    combSet = codeOffCode.merge(combSet, how='right', on=ofc)
        
    combSet = combSet.merge(dataOffCode, how='left', on=ofc)
    combSet[off] = combSet[lit].fillna(combSet[tdo])

    combSet = combSet.drop([lit, tdo], axis='columns')
    combSet[sev] = combSet[cat].fillna('unknown')
    combSet = combSet.drop(cat, axis='columns')
    
    combSet.to_csv('newOffCodes.csv')
    return combSet
```

```python
#read in files
dfOct = pd.read_csv('HVdata-oct.csv')
dfOff = pd.read_excel('offenseCodes.xls')

offenseCodes = writeOffCode(dfOct, dfOff)

#change offense literals in data set to standardize.
dfOct = dfOct.merge(offenseCodes, on=ofc, how='left')
dfOct.drop(toDrop, inplace = True, axis='columns') 
#re arrange
dfOct = dfOct[[ofc, off, sev, sen, sex, rac, age, dat, fac]]
dfOct.dropna(inplace = True)
dfOct[dat] = pd.to_datetime(dfOct[dat])

#reset the indexes
dfOct = dfOct.reset_index()

dfOct.to_csv('data.csv')
senYrs = dfOct[sen]
filter = pd.to_numeric(dfOct[sen], errors='coerce')
filter = np.isnan(filter)
filter = np.asarray(filter.loc[filter].index)

#replace in sentence years so we can have numbers there
sentence = {DEATH : 0, CAPLIFE : -1, LIFE : -2, LWOP : -3}
nonNums = senYrs.take(list(filter))
#replace the words with numbers
for key in sentence.keys():
    nanArray = np.logical_not(np.array(nonNums.str.isnumeric()))
    nonNums[nanArray & nonNums.str.contains(key)] = sentence[key]

senYrs.drop(filter, inplace=True, axis=0)
senYrs = senYrs.combine_first(nonNums)

data = dfOct.copy() #make it easier on myself referencing stuff

data[sen] = senYrs

# write this a csv file, makes it easier to get a good look at
dfOct.to_csv('data.csv')
```
